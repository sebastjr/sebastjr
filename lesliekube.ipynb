# Lab 2 - Leslie kube

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
svart = np.array([11.41, 16.06, 19.04, 20.67])
hvit = np.array([11.06, 15.52, 18.61, 20.34])
matt = np.array([3.45, 4.94, 5.70, 6.30])
skinnende = np.array([0.59, 1.03, 1.38, 1.44])
P = np.array([5, 6.5, 8, 10])

hvit_e = hvit/svart
matt_e = matt/svart
skinnende_e = skinnende/svart

avghvit_e = np.mean(hvit_e)
avgmatt_e = np.mean(matt_e)
avgskinnende_e = np.mean(skinnende_e)

varhvit_e = np.var(hvit_e)
varmatt_e = np.var(matt_e)
varskinnende_e = np.var(skinnende_e)
print('Varians for hvit epsilon ', varhvit_e)
print('Varians for matt epsilon ', varmatt_e)
print('Varians for skinnende epsilon ', varskinnende_e)

plt.figure(figsize=(8, 8))
plt.plot(P, hvit_e, marker = 'o', ls = ' ', label = 'Hvit', color = 'blue')
plt.axhline(avghvit_e, ls = '--', color = 'blue', label = f'Avg. $\epsilon = ${round(avghvit_e, 4)}')

plt.plot(P, matt_e, marker = 'o', ls = ' ', label = 'Metallisk matt', color = 'orange')
plt.axhline(avgmatt_e, ls = '--', color = 'orange', label = f'Avg. $\epsilon = ${round(avgmatt_e, 4)}')

plt.plot(P, skinnende_e, marker = 'o', ls = ' ', label = 'Metallisk skinnende', color = 'green')
plt.axhline(avgskinnende_e, ls = '--', color = 'green', label = f'Avg. $\epsilon = ${round(avgskinnende_e, 4)}')

plt.yticks(np.linspace(0, 1, 11))
plt.yticks(np.linspace(0, 1, 21), minor = True)
plt.xticks([5, 6.5, 8, 10], [5, 6.5, 8, 'high'])

plt.title('Emisjonsevne $\epsilon$ for ulike flater p Leslie kube relativt til svart flate')
plt.ylabel('Emisjonsevne $\epsilon$')
plt.xlabel('Effektinnstilling P')

plt.grid()
plt.legend()

# plt.savefig('lesliekube.pdf', format = 'pdf', bbox_inches="tight")
plt.show()
```
